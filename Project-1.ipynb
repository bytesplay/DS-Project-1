3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# hh.isnull().sum()
# Изначально явных пропусков в "Опыт работы": 168
hh['Опыт работы']=hh['Опыт работы'].apply(lambda value: np.nan if value =='Не указано' else value)
hh.isnull().sum()
# После замены "Не указано" на NaN пропусков стало 170

def month_experience(experience):
    """
    Функция-обработчик. Конвертирует информацию об опыте работы в количество месяцев работы
    """
    try:
        experience = experience.split(' ')
        # Опыт[0] работы[1] 3[2] года[3] 2[4] месяца[5]…
        if experience[5].startswith("месяц"):
            months = int(experience[2])*12 + int(experience[4])
        elif experience[3].startswith("месяц"):
            months = int(experience[2])
        else:
            months = int(experience[2])*12
    except:
        months = np.nan
    return months

hh['Опыт работы (месяц)'] = hh['Опыт работы'].apply(month_experience)
hh.drop('Опыт работы', axis=1, inplace=True)

hh['Опыт работы (месяц)'].median()
# 100.0


```

    * … , готов к командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python


def drop_city_to_move(info):
    """
    Функция-обработчик. Удаляет из записей информацию о городах, желаемых для переезда
    """
    updated_info = info.split('(')[0].strip(' ')
    if info != updated_info:
        addition = info.split(')')[-1].strip(' ')
        updated_info += addition
    return updated_info

def get_city(info):
    """
    Функция-обработчик. Извлекает название города проживания
    """
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск',
                     'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград', 'город-миллионник']
    city = info.split(',')[0].strip(' ')
    if city in ['Москва', 'Санкт-Петербург']:
        pass
    elif city in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
    return city


def get_movement(info):
    """
    Функция-обработчик. Извлекает информацию о возможности переезда
    """
    negative = ['не готов к переезду', 'не готова к переезду']
    return not (info.find(negative[0])!=-1 or info.find(negative[1])!=-1)


def get_trip(info):
    """
    Функция-обработчик. Извлекает информацию о командировках
    """
    negative = ['не готов к командировкам', 'не готова к командировкам']
    mention = 'командировкам'
    # if (info.find(mention)!=-1):
    return not (info.find(negative[0])!=-1 or info.find(negative[1])!=-1)
    # else:
    #     return False
    
hh['Город, переезд, командировки'] = hh['Город, переезд, командировки'].apply(drop_city_to_move)
hh['Город'] = hh['Город, переезд, командировки'].apply(get_city)
hh['Город'].value_counts(normalize=True)

hh['Готовность к переезду'] = hh['Город, переезд, командировки'].apply(get_movement)
hh['Готовность к командировкам'] = hh['Город, переезд, командировки'].apply(get_trip)

# Questions
simult = hh['Готовность к переезду'] & hh['Готовность к командировкам']
hh.drop('Город, переезд, командировки', axis=1, inplace=True)
simult.value_counts(normalize=True)

```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Создание признаков One Hot Encoding
employments = ['полная занятость', 'частичная занятость', 'проектная работа', 'стажировка', 'волонтерство']
for employment in employments:
    hh[employment] = hh['Занятость'].apply(lambda value: False if value.find(employment)==-1 else True)

schedules = ['гибкий график', 'полный день', 'сменный график', 'удаленная работа', 'вахтовый метод']
for schedule in schedules:
    hh[schedule] = hh['График'].apply(lambda value: False if value.find(schedule)==-1 else True)

# Questions
project_vol = hh['проектная работа'] & hh['волонтерство']
shift_flex = hh['гибкий график'] & hh['вахтовый метод']

# project_vol.value_counts()
# True       436

# shift_flex.value_counts()
# True      2311

hh.drop(['Занятость', 'График'], axis=1, inplace=True)
```

```python
xrates = pd.read_csv('data/ExchangeRates.csv', sep=',')

hh['Обновление резюме'] = hh['Обновление резюме'].apply(lambda d: d.split(' ')[0])
# Убираем значение времени
hh['Обновление резюме'] = pd.to_datetime(hh['Обновление резюме'], dayfirst=True)

def get_currency(record):
    currency = {
        # соответствие типов валют на hh.ru и ISO
        'грн.': 'UAH',
        'USD': 'USD',
        'EUR':'EUR',
        'бел.руб.': 'BYN',
        'KGS': 'KGS',
        'сум': 'UZS',
        'AZN': 'AZN',
        'KZT': 'KZT',
        'руб.': 'RUB'
    }
    if record:
        source_currency = record.split(' ')[1]
        # Определяем тип валюты на сайте hh.ru
        return currency[source_currency]
        # возвращаем тип вылюты согласно ISO кодировке
    else:
        pass
    

hh['salary'] = hh['ЗП'].apply(lambda value: int(value.split(' ')[0]))
# Определяем признак со значением зарплаты
hh['salary_currency'] = hh['ЗП'].apply(get_currency)
# Определяем признак с типов валюты ISO
xrates['date'] = pd.to_datetime(xrates['date'], dayfirst=True)

hh = hh.merge(
    xrates,
    left_on=['salary_currency', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
)

hh = hh.fillna( {'close': 1, 'proportion': 1})
# заполнение пропорции для валюты руб. единицами
hh['ЗП (руб)'] = hh['salary']*hh['close'] / hh['proportion']


display(hh['ЗП (руб)'].median())
# 59019.0
del_cols = ['salary', 'salary_currency', 'currency', 'per',
            'date', 'time', 'close', 'vol', 'proportion', 'ЗП']

hh.drop(del_cols, axis=1, inplace=True)
hh.info()
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
def outliers_z_score_mod(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return x, outliers, cleaned, mu, lower_bound, upper_bound

data, outliers, cleaned_data, mean_age, lower_age, upper_age, = outliers_z_score_mod(hh, 'Возраст', left=3,  right=4, log_scale=True)
print('Количество выбросов:',outliers.shape[0])

fig = plt.figure(figsize=(10, 6))
histplt = sns.histplot(
    data,
    bins=30,
    kde=True,
    log_scale=True
);

histplt.axvline(mean_age, color='k', lw=2)
histplt.axvline(lower_age, color='k', lw=2, ls='--')
histplt.axvline(upper_age, color='k', lw=2, ls='--')

histplt.set_title('Логарифмированное распределение соискателей по возрасту');
hh.drop(outliers.index, axis=0, inplace=True)
```
